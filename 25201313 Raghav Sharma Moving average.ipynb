```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mu = 10
phi1 = 0.5 
phi2 = 0.3
n = 20
```

```python
np.random.seed(42) 
epsilon = np.random.randint(-3, 4, size=n)
```

```python
f_t_ma1 = np.zeros(n)
f_t_ma2 = np.zeros(n)
```

```python
for t in range(1, n):
    f_t_ma1[t] = mu + phi1 * epsilon[t-1] # MA(1)
    if t >= 2:
        f_t_ma2[t] = mu + phi1 * epsilon[t-1] + phi2 * epsilon[t-2] # MA(2)
    else:
        f_t_ma2[t] = mu + phi1 * epsilon[t-1]
```

```python
df = pd.DataFrame({
"t": np.arange(1, n+1),
"ε_t": epsilon,
"MA(1)": f_t_ma1,
"MA(2)": f_t_ma2
})
```

```python
print(df)
```

```python
# Plot
plt.figure(figsize=(12,6))


plt.plot(df["t"], df["ε_t"], marker="o", linestyle="--", color="red", label="ε_t (errors)")
plt.plot(df["t"], df["MA(1)"], marker="s", linestyle="-", color="blue", label="MA(1) values")
plt.plot(df["t"], df["MA(2)"], marker="^", linestyle="-", color="green", label="MA(2) values")


plt.axhline(mu, color="gray", linestyle=":", label="mean (μ)")
plt.xlabel("Time (t)")
plt.ylabel("Value")
plt.title("Moving Average Models: MA(1) vs MA(2)")
plt.legend()
plt.grid(True)
plt.show()
```
